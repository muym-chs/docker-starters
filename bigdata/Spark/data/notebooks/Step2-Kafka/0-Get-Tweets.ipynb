```python
import time
import requests
import re
from pprint import pformat
```

```python
def filterData(item) :
    
    cleansed_data = {
        "id" : item['id'],
        "content" : item['content'],
        "sendTime" : item['sendTime'],
        "sendTimePersian" : item['sendTimePersian'],
        "senderName" : item['senderName'],
        "senderUsername" : item['senderUsername'],
        "type" : item['type'],
    } 
    return cleansed_data
```

```python
url = "https://www.sahamyab.com/guest/twiter/list?v=0.1"
delay = 10
# tweet_ids=set()
cnt=0
while True:
    
    response = requests.request('GET', url, headers={'User-Agent': 'Chrome/81'})
    if response.status_code == requests.codes.ok:
        items = response.json()['items']
        for item in items:
            cnt+=1
            print("-=-"*20)
            print(f"#{cnt:4}")
            if not item.get('content') :
                print("tweet is unacceptable")
                continue
#             elif item['id'] in tweet_ids:
#                 print(f"tweet with id : {item['id']} already received.")
#                 continue
            else : 
                print(f"{pformat(filterData(item))}")
#                 tweet_ids.add(item["id"])
    else:
        print("Error in fetch data: {err}".format(err=response.status_code))
    time.sleep(delay)

```
